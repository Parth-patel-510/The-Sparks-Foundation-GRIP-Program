```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data= pd.read_csv('SampleSuperstore.csv')
```

```python
data.drop(['Country','Postal Code'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(15,10))
data['Ship Mode'].value_counts().plot.pie(autopct="%.1f%%")
plt.title('Ship Mode pie chart')
```

<b> the pie chart shows the Ship mode percentage as it is clearly seen the oders mostly belongs to standard class more followed by the second class at second highest percentage 

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Ship Mode',y='Profit',data=data)
plt.title('Profit over Ship Mode')
```

<b>the bar graph shows profit over Ship Mode no such difference here as the values are almost around each other
    
    As we can see that there is no such noticable relation of ship mode with profit as they are all almost identical

```python
plt.figure(figsize=(15,10))
data['Segment'].value_counts().plot.pie(autopct="%.1f%%")
plt.title('Segment pie chart')
```

<b> the above pie chart shows order segment thus the consumers here outweigh more in numbers than the other two

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Segment',y='Profit',data=data)
plt.title('Profit over Segments')
```

<b> however the profits is more from the other two segments of customer 
    
    thus talking about customers Segment we experience more profit in the Home office department followed by the corporate     segment and Consumer segement.

```python
plt.figure(figsize=(15,10))
data['Category'].value_counts().plot.pie(autopct="%.1f%%")
```

<b> talking about the category of the product more number of office supplies are purchased over the other two

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Category',y='Profit',data=data)
plt.title('Profit over Categories')
```

<b> the profit share is more in the technology category

```python
plt.figure(figsize=(15,10))
data['Sub-Category'].value_counts().plot.pie(autopct="%.1f%%")
```

<b> the pie charge above represent the 17 sub category of the product orderd blinders and papers are more in numbers

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Sub-Category', hue='Segment',data=data)
plt.xticks(rotation=90)
plt.show()
```

<b> The consumers outweigh the number of sales over any other segments

```python
plt.figure(figsize=(15,10))
sns.countplot(x='State',data=data)
plt.xticks(rotation=90)
plt.show()
```

<b> the above count plot shows that order from california are more

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Region',data=data)
```

<b> Sales are pretty less in South and central Region

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Region', hue='Segment',data=data)
```

<b>also as previously visualized that Home office has max profit than other two thus area to target must be home office in all four region 

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Discount',y='Sales',data=data)
plt.title('Sales over Discount')
```

<b> figure shows that as the discount increases to 5% the sales increases

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Discount',y='Profit',data=data)
plt.title('Profit over Discount')
```

<b> but the Profit decreases as we increases the discount rate thus we cannot exceed more than 2%
